# What's the most popular programming language in 2020 and 2021?

```python
question7 = question.filter(regex = 'Q7')[0].split("-")[0]
question7
```

```python
question7_20 = question2020.filter(regex = 'Q7')[0].split("-")[0]
question7_20
```

```python
answer7 = answer.filter(regex = 'Q7')
answer7.head()
```

```python
answer7_20 = answer2020.filter(regex = 'Q7')
answer7_20.head()
```

```python
answer7.columns
```

```python
answer7.notnull().sum() # sum all the answers
```

```python
answer7_describe = answer7.describe() 
answer7_describe
```

```python
answer7 = answer7_describe.loc[['top', 'count']].transpose()
answer7 = answer7.sort_values('count', ascending = False)

answer7
```

```python
answer7_20_describe = answer7_20.describe()
answer7_20_describe
```

```python
answer7_20 = answer7_20_describe.loc[['top', 'count']].transpose()
answer7_20 = answer7_20.sort_values('count', ascending = False)

answer7_20
```

```python
result7 = pd.merge(answer7,answer7_20, how='left', on='top')
result7
```

```python
result7.rename(columns = {'count_x': '2021', 'count_y': '2020'}, inplace = True)
```

```python
result7['2021'] =  result7['2021'].astype(float)
```

```python
result7['2020'] =  result7['2020'].astype(float)
```

```python
result7.info()
```

```python
def percentage_change(current, previous):
    if previous != 0 :
        return float(current - previous) / abs(previous) * 100
    else:
        return "undefined"
```

```python
result7['change_rate'] = result7.apply(lambda x: percentage_change(x['2021'], x['2020']), axis=1)
```

```python
result7
```


**now i will use barplot from seaborn to visualiza the data and compare it**
**and plot it as well the rate change**


```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5), sharey=True)
fig.suptitle('The most popular programming language in 2020 and 2021', fontsize = 12, color = 'lightpink', weight = 'bold')



sns.barplot(ax=axes[0], data = answer7_20, x='count', y = 'top', color='#ffab40')
axes[0].set_title('2020')
axes[0].grid(axis='x', lw = 0.25)

sns.barplot(ax=axes[1], data = answer7, x='count', y = 'top', color='#ff4094')
axes[1].set_title('2021')
axes[1].grid(axis='x', lw = 0.25)


plt.savefig('plot3.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

```python
result7.plot(x='top', y=['2020','2021'], kind='bar', color=['#ff4094','#ffab40'])
plt.title('The most popular programming language in 2020 and 2021', pad = 30, fontsize = 12, color = 'lightpink', weight = 'bold')
plt.xticks(rotation=45)
plt.grid(axis='y', lw = 0.25)

plt.savefig('plot4.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

```python
result7.plot(x='top', y=['change_rate'], kind='bar', color=['#ffab40']);
plt.title('change rating', pad = 30, fontsize=10, color='lightpink',weight = 'bold');
plt.xticks(rotation=45)
plt.grid(axis='y', lw = 0.25)
plt.savefig('plot5.png', dpi=300, bbox_inches='tight');
plt.show();
```
